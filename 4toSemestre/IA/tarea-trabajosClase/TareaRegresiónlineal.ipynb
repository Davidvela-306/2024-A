```python
import pandas as pd
datos=pd.read_csv("seguro_salud.csv")
print(datos.head())
print(datos.describe())
#separar caracterisiticas (x) y las salidas (y)
x=datos.iloc[:,:-1] #todas las caracteristicas
y=datos.iloc[:,-1]

from sklearn.model_selection import train_test_split

#dividir los datos de conjuntos de entrenamientos y prueba
x_train, x_test, y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state=70)
from sklearn.linear_model import LinearRegression
# crear modelo de regresion lineal
modelo = LinearRegression()
#entrenar el modelo
modelo.fit(x_train,y_train)

#Evaluar el modelo con los datos de prueba
y_obt=modelo.predict(x_test)
from sklearn.metrics import mean_squared_error,r2_score

#metricas de evaluacion
error=mean_squared_error(y_test,y_obt)
print("Error cuandratico medio es:",error)
r2 =r2_score(y_test,y_obt)
print("el coeficiente de determinacion es:",r2)


#realizar prediciion
y_obt=modelo.predict([[55,1,35,2,1]])
print("La Predicción de los cargos del seguro es: ",y_obt)
#Predicciones de la tarea:
print("**********************************+")
y_pred = modelo.predict([[30, 0, 30.0, 0, 0]])
print("La Predicción de los cargos del seguro es: ", y_pred)

y_pred = modelo.predict([[42, 1, 45.0, 3, 1]])
print("La Predicción de los cargos del seguro es: ", y_pred)

y_pred = modelo.predict([[55, 0, 35.0, 2, 1]])
print("La Predicción de los cargos del seguro es: ", y_pred)
print("**********************************+")

#coeficientes  e intercepcion
print("coeficientes :",modelo.coef_)
print("intercepcion :",modelo.intercept_)

#Gráfica de dispersion
import matplotlib.pyplot as plt

# "age" como variable independiente (x)
x_age = datos["age"].values.reshape(-1, 1)

# "cargos" como variable dependiente (y)
y_charges = datos["charges"].values


plt.scatter(x_age, y_charges)
plt.xlabel("Edad")
plt.ylabel("Costo del seguro")
plt.title("Gráfica ")
plt.show()
```
